```python
import numpy as np
import networkx as nx
import torch
from torch_geometric.data import Data


def graph_to_gcn_input(graph):

    node_features = []
    node_mapping = {}

    for idx, (node, attr) in enumerate(graph.nodes(data=True)):

        features = [
            attr['roomType_embd'],  # Room type embedding
            attr['actualCentroid_x'],
            attr['actualCentroid_y'],
            attr['rec_w'],
            attr['rec_h'],
            attr['roomSize']
        ]
        node_features.append(features)
        node_mapping[node] = idx


    X = torch.tensor(node_features, dtype=torch.float)


    adj_matrix = np.zeros((len(graph.nodes()), len(graph.nodes())), dtype=int)

    for edge in graph.edges(data=True):
        node1, node2, _ = edge
        idx1 = node_mapping[node1]
        idx2 = node_mapping[node2]
        adj_matrix[idx1, idx2] = 1
        adj_matrix[idx2, idx1] = 1


    edge_index = torch.tensor(np.array(np.nonzero(adj_matrix)), dtype=torch.long)


    data = Data(x=X, edge_index=edge_index)

    return data


first_graph = planify_graphs[0]
gcn_data = graph_to_gcn_input(first_graph)

print(f"Node features shape: {gcn_data.x.shape}")
print(f"Edge index shape: {gcn_data.edge_index.shape}")
```

```python
import pickle
import torch
import networkx as nx
from torch_geometric.data import Data, DataLoader
from torch_geometric.nn import GCNConv


def load_graphs(pickle_file):
    with open(pickle_file, 'rb') as f:
        return pickle.load(f)


graphs_all_connected = load_graphs('/content/drive/MyDrive/Planify_Graphs_Scaled/Graphs_all_conected.pkl')
graphs_living_to_all = load_graphs('/content/drive/MyDrive/Planify_Graphs_Scaled/Graphs_living_to_all.pkl')
graphs_real = load_graphs('/content/drive/MyDrive/Planify_Graphs_Scaled/Graphs_real.pkl')
boundaries = load_graphs('/content/drive/MyDrive/Planify_Graphs_Scaled/boundaries.pkl')


def extract_features_and_edges(graphs):
    node_features = []
    edge_index = []
    labels = []

    for graph in graphs:
        if not isinstance(graph, nx.Graph):
            continue

        for node, data in graph.nodes(data=True):

            features = [
                data.get('roomType_embd', -1),
                data.get('actualCentroid_x', 0.0),
                data.get('actualCentroid_y', 0.0),
                data.get('rec_w', 0.0),
                data.get('rec_h', 0.0)
            ]
            node_features.append(features)


            labels.append(data.get('roomType_embd', -1))

        for u, v in graph.edges():
            if isinstance(u, int) and isinstance(v, int):
                edge_index.append((u, v))


    edge_index = torch.tensor(edge_index, dtype=torch.long).t().contiguous() if edge_index else torch.empty((2, 0), dtype=torch.long)
    node_features = torch.tensor(node_features, dtype=torch.float) if node_features else torch.empty((0, 5), dtype=torch.float)
    labels = torch.tensor(labels, dtype=torch.float) if labels else torch.empty((0,))  # Labels as a tensor

    return node_features, edge_index, labels


all_node_features = []
all_edge_index = []
all_labels = []


for graphs in [graphs_all_connected, graphs_living_to_all, graphs_real, boundaries]:
    node_features, edge_index, labels = extract_features_and_edges(graphs)
    all_node_features.append(node_features)
    all_edge_index.append(edge_index)
    all_labels.append(labels)


combined_node_features = torch.cat(all_node_features, dim=0) if all_node_features else torch.empty((0, 5), dtype=torch.float)
combined_edge_index = torch.cat(all_edge_index, dim=1) if all_edge_index else torch.empty((2, 0), dtype=torch.long)
combined_labels = torch.cat(all_labels, dim=0) if all_labels else torch.empty((0,))  # Combine all labels

data = Data(x=combined_node_features, edge_index=combined_edge_index, y=combined_labels)


from torch_geometric.loader import DataLoader
dataloader = DataLoader([data], batch_size=32, shuffle=True)


class GCNModel(torch.nn.Module):
    def __init__(self, input_dim, output_dim):
        super(GCNModel, self).__init__()
        self.conv1 = GCNConv(input_dim, 64)
        self.conv2 = GCNConv(64, output_dim)

    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        x = self.conv1(x, edge_index)
        x = torch.nn.functional.relu(x)
        x = self.conv2(x, edge_index)
        return x


gcn_model = GCNModel(input_dim=combined_node_features.shape[1], output_dim=1)
loss_fn = torch.nn.MSELoss()
optimizer = torch.optim.Adam(gcn_model.parameters(), lr=0.01)


def train_model(model, dataloader, optimizer, loss_fn, num_epochs=20):
    model.train()
    for epoch in range(num_epochs):
        for data in dataloader:
            optimizer.zero_grad()
            out = model(data)
            if hasattr(data, 'y') and data.y is not None:
                loss = loss_fn(out.view(-1), data.y)
                loss.backward()
                optimizer.step()
            else:
                print("Warning: data.y is not defined or is None.")
        print(f'Epoch {epoch + 1}/{num_epochs}, Loss: {loss.item()}')


train_model(gcn_model, dataloader, optimizer, loss_fn, num_epochs=20)


torch.save(gcn_model.state_dict(), 'gcn_model.pth')
```

```python
import matplotlib.pyplot as plt
import networkx as nx
from torch_geometric.data import Data

def visualize_graph(graph, model, device):
    """Visualizes a given graph using the GCN model to predict node embeddings."""


    model.eval()


    node_features, edge_index, _ = extract_features_and_edges([graph])
    data = Data(x=node_features, edge_index=edge_index)


    data = data.to(device)


    with torch.no_grad():
        predictions = model(data).cpu().numpy()


    pos = nx.spring_layout(graph)


    plt.figure(figsize=(10, 8))
    nx.draw(graph, pos, with_labels=True, node_size=700, node_color=predictions.flatten(), cmap=plt.cm.viridis)
    plt.title("Graph Visualization with GCN Predictions")
    plt.colorbar(plt.cm.ScalarMappable(cmap=plt.cm.viridis), label='Predicted Room Type Embedding')
    plt.show()


graph_to_visualize = graphs_all_connected[0]


device = 'cuda' if torch.cuda.is_available() else 'cpu'
visualize_graph(graph_to_visualize, gcn_model, device)
```

```python
import matplotlib.pyplot as plt
import networkx as nx

def visualize_architectural_drawing(graph):
    plt.figure(figsize=(12, 10))


    min_x, max_x = float('inf'), float('-inf')
    min_y, max_y = float('inf'), float('-inf')


    for node, data in graph.nodes(data=True):

        width = data.get('rec_w', 5)
        height = data.get('rec_h', 5)
        x = data.get('actualCentroid_x', 0)
        y = data.get('actualCentroid_y', 0)


        min_x = min(min_x, x - width / 2)
        max_x = max(max_x, x + width / 2)
        min_y = min(min_y, y - height / 2)
        max_y = max(max_y, y + height / 2)


        plt.gca().add_patch(plt.Rectangle((x - width / 2, y - height / 2), width, height, fill=None, edgecolor='black', linewidth=2))


        plt.text(x, y, str(node), horizontalalignment='center', verticalalignment='center')


    for u, v in graph.edges():
        x_u, y_u = graph.nodes[u]['actualCentroid_x'], graph.nodes[u]['actualCentroid_y']
        x_v, y_v = graph.nodes[v]['actualCentroid_x'], graph.nodes[v]['actualCentroid_y']
        plt.plot([x_u, x_v], [y_u, y_v], color='grey', linewidth=2)


    plt.xlim(min_x - 5, max_x + 5)
    plt.ylim(min_y - 5, max_y + 5)

    plt.gca().set_aspect('equal', adjustable='box')
    plt.title("Architectural Floor Plan Visualization")
    plt.xlabel("X Coordinate")
    plt.ylabel("Y Coordinate")
    plt.grid(True)
    plt.show()


graph_to_visualize = graphs_all_connected[0]
visualize_architectural_drawing(graph_to_visualize)
```
